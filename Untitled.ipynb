```python
import pandas as pd
import numpy as np
from pathlib import Path

pd.set_option('display.max_columns', None)   
pd.set_option('display.width', 200)           
```

```python
RAW = Path("data/raw")

files = {
    "customers":   "olist_customers_dataset.csv",
    "orders":      "olist_orders_dataset.csv",
    "order_items": "olist_order_items_dataset.csv",
    "payments":    "olist_order_payments_dataset.csv",
    "reviews":     "olist_order_reviews_dataset.csv",
    "products":    "olist_products_dataset.csv",
    "sellers":     "olist_sellers_dataset.csv",
    "geolocation": "olist_geolocation_dataset.csv",
    "category":    "product_category_name_translation.csv",
}

data = {name: pd.read_csv(RAW / fname) for name, fname in files.items()}

for name, df in data.items():
    print(f"  {name:12s} -> {df.shape}")
```

```python
for name, df in data.items():
    miss = df.isnull().sum()       
    miss = miss[miss > 0]          
    for col, n in miss.items():
        pct = round(n / len(df) * 100, 2)
        print(f"{name} | {col} | {n} ({pct}%)")
```

Data Dictionary

```python
rows = []
for table, df in data.items():
    n = len(df)
    for col in df.columns:
        miss = df[col].isnull().sum()
        non_null = df[col].dropna()
        sample = str(non_null.iloc[0]) if len(non_null) else ""
        rows.append({
            "table": table, "column": col, "dtype": str(df[col].dtype),
            "missing_pct": round(miss/n*100, 2),
            "n_unique": df[col].nunique(), "sample_value": sample,
        })
dd = pd.DataFrame(rows)
dd
```
